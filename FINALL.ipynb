```python
import pandas as pd
```

### Delhi Housing Prices
### 0: Absent
### 1: Present

```python
Del=pd.read_csv('Delhi3.csv')
```

```python
Del
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, ax = plt.subplots(figsize=(30,30))         # Sample figsize in inches
sns.heatmap(Del.corr(), annot=True, ax=ax)
```

### Here we can see that the Wifi and the Golfcourse columns are completely empty so we can drop these values.

```python
Del=Del.drop("Wifi",axis=1)
Del=Del.drop("GolfCourse",axis=1)
Del=Del.drop("Wardrobe",axis=1)
```

```python
Del.describe()
```

```python
Del2=Del.copy()
```

```python
Del2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
Del2['Location']=le.fit_transform(Del2['Location'])
```

```python
Del2.head()
```

```python
X=Del2.drop('Price',axis=1)
X
```

```python
y=Del2['Price']
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=20)
```

### Using Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model = RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print(y_pred)
```

### Check Accuracy for Random Tree Regression

```python
from sklearn.metrics import r2_score
r2_score(y_test,y_pred)
```

### Using Decision Tree Regressor

```python
from sklearn.tree import DecisionTreeRegressor
model2=DecisionTreeRegressor()
```

```python
model2.fit(X_train,y_train)
```

```python
y_pred2 =  model2.predict(X_test)
```

### Checking Accuracy for Decision Tree Regression

```python
r2_score(y_test,y_pred2)
```

### Using Ridge Regressor

```python
from sklearn.linear_model import Ridge
model3 = Ridge()
```

```python
model3.fit(X_train,y_train)
```

```python
y_pred3=model3.predict(X_test)
```

### Checking Accuracy for Ridge Regression

```python
r2_score(y_test,y_pred3)
```

### Using Linear Regression

```python
from sklearn.linear_model import LinearRegression
model4=LinearRegression()
```

```python
model4.fit(X_train,y_train)
```

```python
y_pred4=model4.predict(X_test)
```

### Checking Accuracy for Linear Regression

```python
r2_score(y_test,y_pred4)
```
